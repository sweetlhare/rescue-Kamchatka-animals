```python
import torch
import cv2


# @st.cache()
def load_model(path='models/best.pt'):
    detection_model = torch.hub.load('ultralytics/yolov5', 'custom', path=path)
    return detection_model


def detect_image(path, model):
    
    image = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)
    pred = detection_model(image)
    pred_df = pred.pandas().xyxy[0].sort_values('confidence', ascending=False)
    pred_image = pred.render()[0]
    
    if pred_df.shape[0] > 0:
        if pred_df.confidence.iloc[0] > 0.25:
            return pred_image, pred_df.name.iloc[0]
        else:
            return pred_image, '–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∂–∏–≤–æ—Ç–Ω–æ–µ'
    else:
        return image, '–ù–µ—Ç –∂–∏–≤–æ—Ç–Ω–æ–≥–æ'
    
    
detection_model = load_model('yolov5/runs/train/exp3/weights/best.pt')
image, prediction = detect_image('yolo_data/val/images/001_396.jpg', detection_model)
```
